```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Generate synthetic Superstore sales data
np.random.seed(42)
num_rows = 500

categories = ['Furniture', 'Office Supplies', 'Technology']
sub_categories = {
    'Furniture': ['Chairs', 'Tables', 'Bookcases'],
    'Office Supplies': ['Paper', 'Binders', 'Pens'],
    'Technology': ['Phones', 'Accessories', 'Copiers']
}
regions = ['East', 'West', 'Central', 'South']
segments = ['Consumer', 'Corporate', 'Home Office']

data = []
base_date = datetime(2023, 1, 1)

for i in range(num_rows):
    cat = random.choice(categories)
    sub_cat = random.choice(sub_categories[cat])
    order_date = base_date + timedelta(days=random.randint(0, 365))
    data.append({
        'Order ID': f"ORD{i+1000}",
        'Order Date': order_date.strftime('%Y-%m-%d'),
        'Region': random.choice(regions),
        'Category': cat,
        'Sub-Category': sub_cat,
        'Sales': round(np.random.uniform(100, 2000), 2),
        'Profit': round(np.random.uniform(-200, 500), 2),
        'Quantity': np.random.randint(1, 10),
        'Customer Segment': random.choice(segments)
    })

df = pd.DataFrame(data)
df.to_csv('superstore_data.csv', index=False)
print("✅ superstore_data.csv generated!")
df.head()
```
